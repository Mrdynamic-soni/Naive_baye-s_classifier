```python
#label encoding without using label encoder library
new_df['sentiment'].replace({'positive':1,'negative':0},inplace = True)
```

```python
# just a check code 
import re 
clean = re.compile('<.*?>')  # to look for the pattern on the text
re.sub(clean,'',new_df.iloc[0].review)
# sun takes three arguements pattern that has to searched and with what should it be
# replaced and where to apply rest pf two
```

```python
# function to clean data of reveiews
def clean_reveiew(text):
    clean = re.compile('<.*?>')
    return re.sub(clean,'',text)
```

```python
new_df['review']=new_df['review'].apply(clean_reveiew)
new_df
```

```python
def make_lower(text):
    return text.lower()
```

```python
new_df['review']=new_df['review'].apply(make_lower)
new_df
```

```python
new_df
```

```python
new_df.shape
```

```python
def remove_special_character(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
new_df['review']  = new_df['review'].apply(remove_special_character)
```

```python
new_df
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
```

```python
def remove_stopwords(text):
    temp = []
    for word in text.split():
        if word not in stop:
            temp.append(word)
    return temp
```

```python
new_df["review"]=new_df["review"].apply(remove_stopwords)
new_df
```

```python
# for stemming
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
y=[]
def stem_word(text):
    for word in text:
        y.append(ps.stem(word))
    x=y[:]
    y.clear()
    return x
```

```python
new_df["review"]=new_df["review"].apply(stem_word)
```

```python
new_df
```

```python
def join_back(text):
    return " ".join(text)
```

```python
new_df["review"]=new_df["review"].apply(join_back)
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(max_features=1000)
```

```python
temp = cv.fit_transform(new_df['review']).toarray()
```

```python
temp.shape
```

```python
out = new_df.iloc[:,-1].values
out
```

```python
out.shape
```

```python
# split data in training set and testing set
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(temp,out, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(xtrain,ytrain)
clf2.fit(xtrain,ytrain)
clf3.fit(xtrain,ytrain)
```

```python
y_pred1 = clf1.predict(xtest)
y_pred2 = clf2.predict(xtest)
y_pred3 = clf3.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Gaussian",accuracy_score(ytest,y_pred1))
print("Multinomial",accuracy_score(ytest,y_pred2))
print("Bernauli",accuracy_score(ytest,y_pred3))
```
